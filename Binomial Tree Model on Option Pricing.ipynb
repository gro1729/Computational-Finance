The Binomial Model: 
====================

If I know for sure that the value of an option at expiration would be $X$, then I would assert that the current value of the option exactly the current value of payment of X at expiration: $$V(S_{0},t)=e^{-r(T-t)}X$$ $t$ is the current time, $S_{0}$ is the current spot price. So we could write in an abstract way.

If we don’t know the value of the option at expiration. It is a random quantity (i.e., a random variable). $$V(S_{0},t)=e^{-r(T-t)}E[P]$$ where $P$ is the payoff function.

Vanilla, European option, $P$ is a function of spot only. In this case the expectation $E[P]$ is essentially an integral of the payoff function against the probability density function $\mu(S,K,S_{0},T-t,r,\sigma)$

$$V(S_{0},t)=e^{-r(T-t)}E[P]=\int_{0}^{\infty}P(S)\mu(S,K,S_{0},T-t,r,\sigma)dS$$ The Black-Scholes function

Model for the underlying. 
==========================

Starting with the spot price $S$ at time $t$, we allow it to move to either $uS$ or $dS$ in the time frame of length $\Delta t$. To guarantee that the number of states is controlled, we require the $ud=1$. We may then write $u=e^{\sigma\sqrt{\Delta t}}$, $d=e^{-\sigma\sqrt{\Delta t}}$. This is because, $$\frac{\Delta S}{S}=\mu\Delta t+\sigma\sqrt{\Delta t}Z$$ when $Z$ is random, $\Delta S=S(t+\Delta t)-S(t)$, $S=S(t)$, $Z\sim N(0,1^{2})$

Since $\frac{S(t+\Delta t)-S(t)}{S(t)}=\ln(\frac{S(t+\Delta t)}{S(t)})=\mu\Delta t+\sigma\sqrt{\Delta t}Z$

It then gives $$S(t+\Delta t)=S(t)e^{\mu\Delta t+\sigma\sqrt{\Delta t}Z}$$ $\sqrt{\Delta t}\gg\Delta t$, $$S(t+\Delta t)=S(t)e^{\sigma\sqrt{\Delta t}Z}$$ The final simplification is to restrict the values of $Z$ to $\pm1$.

In general, at time $t_{k}$, there are prices $S_{k}^{i}=u^{i}d^{k-i}S_{0}$.

Option Price Tree and Simple European Options
---------------------------------------------

We analyze the relationship between values time $t$ and time $t+\Delta t$ under following assumption:

-   The option is part of the portfolio that is constructed to be risk free.

-   Arbitrage is not allowed.

Consider a portfolio consisting of $\Delta$ shares of the underlying and short one option. Its value at time $t$ is: $$\Pi=\Delta S-V$$ At time $t+\Delta t$, the stock moves either $uS=S_{u}$ or $dS=S_{d}$, The option has corresponding moves to $V_{u}$ or $V_{d}$, and the portfolio then moves to either $\Pi_{u}$ or $\Pi_{d}$. We want to find the value of $\Delta$ that will insure the portfolio has no risk, that is $\Pi_{u}=\Pi_{d}$, which means, $\Delta S_{u}-V_{u}=S_{d}-V_{d}$, thus, $$\Delta=\frac{V_{u}-V_{d}}{S_{u}-S_{d}}\approx\frac{\partial V}{\partial S}$$ We now consider how the risk-free portfolio must grow in a world that does not allow arbitrage. Since it is risk-free, it must grow with rate $r$. otherwise there will be risk-free profit. In which it means, $\Pi(t)=e^{-r\Delta t}\Pi(t+\Delta t)=e^{-r\Delta t}\Pi_{u}$

Which means, $\Delta S-V=e^{-r\Delta t}(uS\Delta-V_{u})$ $$V=e^{-r\Delta t}[V_{u}q+V_{d}(1-q)]$$ where $q=\frac{e^{-r\Delta t}-d}{u-d}$ which can be viewed as a probability measure

Algorithm for option pricing
============================

European Option Pricing
-----------------------

1.  Choose number of time steps $M$ and set $\Delta t=\frac{T}{M}$

2.  Define $u=e^{\sigma\sqrt{\Delta t}}$, $d=\frac{1}{u}$, and $p=\frac{e^{r}\Delta t-d}{u-d}$

3.  Create the values of $S$ at expiration by $S_{M}^{j}=S_{0}u^{j}d^{M-j}$

4.  Evaluate the payoff at expiration to get the option values: $V_{M}^{j}=P(S_{M}^{j})$

5.  For $i=M-1,\cdots,0$ and $j=0,1,\cdots,i$ define: $$V_{i}^{j}=e^{-r\Delta t}(pV_{i+1}^{j+1}+(1-p)V_{i+1}^{j})$$



```python
from math import exp,sqrt
import numpy as np
import matplotlib.pyplot as plt


def payoff(X,strike):
    return np.maximum(0,strike - X)
    
    
def plotBinomial(X):
    for i in range(X.shape[0]):
        plt.plot(np.ones((i+1,1))*i,X[:(i+1),i],'.')

strike = 50;
spot = 50;
r = 0.1 # rate
vol = 0.4
expiry = 5.0/12
nStep = 16;
dt = expiry/nStep
up = exp(vol*sqrt(dt))
down = 1.0/up
q = (exp(r*dt) - down)/(up - down)
disc = exp(-r * dt)


stock = np.zeros((nStep+1,nStep+1))
stock[0,0] = spot

for i in np.arange(1,nStep+1):
    stock[0,i] = stock[0,i-1]*down
    stock[1:(i+1),i] = stock[0:(i),i-1]*up

plotBinomial(stock);
plt.xlabel('time');
plt.ylabel('price');
plt.title('stock');
```

American Option Pricing: 
-------------------------

1.  Choose number of times $M$ and set $\Delta t=\frac{T}{M}$

2.  Define $u=e^{\sigma\sqrt{\Delta t}},d=e^{-\sigma\sqrt{\Delta t}}$

3.  Create the values of $S$ at time $t_{i}$ by $S_{i,j}=S_{0}u^{i}d^{i-j}$ for $0\leq j\leq i$.

4.  Evaluate the payoff at expiration to get the option values: $V_{M}^{j}=sP(S_{M}^{j})$

5.  For $i=M-1,\cdots,0$, and $j=0,1\cdots,i$

    1.  define $tmp=e^{-r\Delta t}(pV_{i+1}^{j+1}+(1-p)V_{i+1}^{j})$

    2.  define $V_{i}^{j}=\max\{temp,P(S_{i}^{j})\}$


```python

# European Option pricing: 
EuroOpt = np.zeros((nStep+1,nStep+1))
EuroOpt[:,-1] = payoff(stock[:,-1],strike)

for i in np.arange(nStep-1,-1,-1):
    for j in range(i+1):
        EuroOpt[j,i] = disc * (q * EuroOpt[j+1,i+1] + (1 - q) * EuroOpt[j,i+1])
        
plotBinomial(EuroOpt);
plt.xlabel('time');
plt.ylabel('price');
plt.title('European option');

```

```python
# American Option pricing: 
AmeOpt = np.zeros((nStep+1,nStep+1))
AmeOpt[:,-1] = payoff(stock[:,-1],strike)

for i in np.arange(nStep-1,-1,-1):
    for j in range(i+1):
        tmp = disc * (q * AmeOpt[j+1,i+1] + (1 - q) * AmeOpt[j,i+1])
        AmeOpt[j,i] = max(tmp,payoff(stock[j,i],50))
        
plotBinomial(AmeOpt);
plt.xlabel('time');
plt.ylabel('price');
plt.title('American option');
```
