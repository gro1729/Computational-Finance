# Deterministic Interest Rate Modeling: 

## Simple Interest
The simplest investment are those whose return is a fixed number - *the interest rate*
, *Intereest earned* on a rincipal amount of *P* invested over the time period *t* if the annual rate of return is $r$
Thus, we have

$$ I = Prt $$

after one year, 

$$P + I = P(1 + r) $$


## Continuous Compounding: 
If the interest is compounded over $N$ time intervals, all of duration $\Delta t$, 
$$\lim_{N\rightarrow \infty} P(1+\frac{r}{N})^N = P e^r $$ 

Or via ordinary differential equation 

$$ \frac{d P}{dt} = rP $$

# A model for random return
The daily returns on stocks cannot be modeled with the ideas discussed above. 

## A General Stochastic Model: 

Let $S = S(t)$ denote the quantity being modeled. A general model of changes in $S$ can be written on 

$$ dS = f(S,t)dt + g(S,t)dR$$

where $g(S,t)$ is some deterministic function and $dR$ captures the randomnes of change in the small time interval $dt$

## The Wiener Process

- W(0) = 0. Convenient normalization
- If $0 < s < u < t$ , then the increments $W(t) - W(u)$ and $W(u) - W(s)$ are independent random variables
- If $s < t$, then $W(t) - W(s)$ is Normally distributed with mean 0 and variance $t - s$
- $W(t)$ is continuous (almost surely)



Thus, this can be modeled as 

- Initialize: $x_0 = 0$ and $y_0 = 0$
- $z ~ N(\mu,\sigma^2)$
- $x_i = x_{i-1} + step $ and $ y_i = y_{i-1} + z /sqrt{step}$


### A example of wiener process is shown below:  

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the main function 
def Wiener():
    # Time step
    n = 100 # Number of end-points
    T = 1.0   # Length of [0,T] in time units
    Delta = T/n
    # Create a vector as for x-axis
    x = np.arange(1, step=Delta)
    # Create an empty vector W of the desired length
    W = np.zeros(n, np.dtype(float)) # For memory-management

    # Z variable, N(0,1)
    mu, sig = 0, 1 # Mu, sigma 

    # The simulated path
    for i in range(1, len(W)):
        W[i] = W[i-1] + np.random.normal(mu, sig) * np.sqrt(Delta)
    plt.plot(x, W)
    return

# Function call
for fcall in range(1,100):
    Wiener()

# Add title
plt.title('Wiener process - 100 paths')

# Display the chart
plt.show()
```

## The Ito Process: 

The Ito Process is a stochastic process $X(t)$, whose infinitesmall increments, $dX$ satisfy the equation 

$$dX = a(X,t)dt+b(X,t)dW$$

where $a$ and $b$ are continuous functions and $dW$ is a Wiener Process. when we say $X$ is a stochastic process, we mean at each $t$, $X(t)$ is a random variable.
